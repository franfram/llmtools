So the basic way in which we use LLM's to perform search over documents is we compare how similar the user's query is to each chunk of text we have from our documents (i.e., we perform something like cosine similarity between the vector that represents the user's query and each vector that represents a chunk of text we have).

The problem with this approach is that sometimes the user query is not very relatable to the chunk of text the LLM actually needs to provide an answer. So when we perform a similarity search (e.g., cosine similarity) to find the most appropriate chunks of text, we end up retrieving chunks that are not very useful. 

One way to tackle this issue is to use Hypotetical Document Embeddings
